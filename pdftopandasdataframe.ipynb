```python
p = figure(plot_height = 600, plot_width = 600, 
           title = 'BTC',
          x_axis_label = 'DATES]', 
           y_axis_label = 'PRICES')

# Add a quad glyph
p.quad(bottom=0, top=data['Date'], 
       left=data['Price'], right=data['Open'], 
       fill_color='red', line_color='black')

# Show the plot
show(p)
```
